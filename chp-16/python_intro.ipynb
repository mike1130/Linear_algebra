## 2B. Methods as functions

A _method_ is a function that is called directly on a variable. 

The list data type, e.g., has a method called _append_ that adds an extra element to and existing list.

```python
aSmallList = ['one', 'more']
print(aSmallList)

aSmallList.append('time')
print(aSmallList)
```

_str_ variables do not have the method _append_

## 2C. Writing your own functions

Creating your own functions is easy and convinient; 

* 1- Use the built-in keyword _def_ to define the function.
    * A _keyword_ is a reserved name that cannot be redefined as a variable or function
* 2- State the function name
* 3- State any possible inputs
* 4- End with a colon ":"

Any lines thereafter are included in the function _if they are indented with two spaces_

Any _outputs_ are indicated by the _return_ keyword


```python
def add2numbers(n1, n2):
    total = n1 + n2
    print(total)
    return total

s = add2numbers(4, 9)
print(s+1)
```

### 3A1. NumPy array

It stores only numbers and have attributes that are useful for mathematical coding.

### 3A2. Indexing and slicing in NumPy

Let's use the function _np.arange(start, end)_ to create and NumPy array which will end at (stop - 1), i.e., Python often uses _exclusive_ upper bounds.

```python
ary = np.arange(-4, 5)
print(ary)
print(ary[5])
```

__Slicing__ is to access more than one element in the NumPy array. It also works with the exclusive upper bound.

```python
# [start: stop]
ary[0:3]
```

__Indexing__ is to access every x element using a defined step.

```python
# [start: stop: skip]
ary[0:5:2]
```

# 5. Translating formulas to code

$$ y = x^2 $$

```python
x = np.arange(-4, 5)
y = x ** 2
plt.plot(x, y, 's-')
```

```python
# [start, stop (inclusive bound), evenly spaced between start-stop]
x = np.linspace(-4, 4, 42)
y = x ** 2 
plt.plot(x, y, 's-')
```

## 5A. Soft coding
Creating variables for parameters that might be change later.

$$
f(x) = \frac{\alpha}{1 + e^{-\beta x}} \\
\quad \alpha = 1.4\\
\quad \beta = 2
$$

```python
x = np.linspace(-4, 4, 42)
alpha = 1.4
beta = 2

num = alpha
den = 1 + np.exp(-beta * x)
fx = num / den
plt.plot(x, fx, 's-')
```

# 7. Control flow

## 7A. Comparators

Special characters that allow to compare different values, which output data type is called a _Boolean_

```python
print(4<5)
print(4>5)
print(4==5)
```

## 7B. If statements

Python will allways proceed from top to bottom. If a conditional is true, Python will run the code associated with the true conditional and skip all following ones.

```python
var = 4
if var == 4:
    print(f'{var} equals 4')
print('Outside the if statement')
```

```python
var = 4

if var == 5:
    print('var is 5')
elif var > 5:
    print('var > 5')
else:
    print('var < 5')
    
print('Outside the if-elif-else')
```

## 7C. For loops

Repeating a code in Python is done through loops.

To create a for loop, you specify an iterable (an iterable is a variable used to iterate over each element in that variable; a list can be used as an iterable)

```python
for i in range(0, 10):      # start, stop (exclusive upper bound)
    print(i+1)
```

```python
theList = [2, 'hello', np.linspace(0, 1, 14)]

for item in theList:
    print(item)
```
